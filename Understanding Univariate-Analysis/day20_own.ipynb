```python
import pandas as pd
import seaborn as sns
```

# 1. Categorical Data

```python
df = sns.load_dataset('geyser')
```

```python
df.head()
```

## a. Countplot

```python
sns.countplot(df['kind'])
```

## b. PieChart

```python
df['kind'].value_counts().plot(kind='pie',autopct='%.2f')
```
